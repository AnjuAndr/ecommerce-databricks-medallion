----------------------------
Notebook: 04_gold_profit_aggregations  
Layer   : Gold  
Purpose : Create analytical profit aggregates for reporting and insights.

Design Decisions:
- Gold layer contains only aggregated, analytics-ready data
- All aggregations are derived from Silver orders to ensure consistency

---------------------



```python
spark.sql("CREATE SCHEMA IF NOT EXISTS gold")
```

```python
#Read Silver Orders
orders_silver = spark.table("silver.orders_enriched")
```

```python
# Aggregate Profit metrics
# Grouping is aligned with reporting requiremnets and business dimensions


from pyspark.sql.functions import sum, round

profit_agg_df = (
    orders_silver
    .groupBy("order_year", "category", "sub_category", "customer_name")
    .agg(
        round(sum("profit"), 2).alias("total_profit")
    )
)
```

```python
#Writing Gold Aggregate Table
profit_agg_df.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("gold.profit_aggregated")
```

```python
#Final validation: confirm Gold schema and tables

spark.table("gold.profit_aggregated").printSchema()
spark.table("gold.profit_aggregated").count()
```
